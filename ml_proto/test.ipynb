```python
# PROGRAM PERHITUNGAN HURUF VOKAL

def hitung_vokal(teks):
  vokal = "aiueoAIUEO"
  jumlah = 0
  for huruf in teks:
    if huruf in vokal:
      jumlah += 1
  return jumlah

teks = input("")
print("Jumlah huruf vokal:", hitung_vokal(teks))
```

```python
# ANAGRAM ATAU BUKAN

def anagram(word1, word2):
  return sorted(word1) == sorted(word2)

word1 = input("Masukkan kata pertama Anda: ")
word2 = input("Masukkan kata kedua Anda: ")

if anagram(word1, word2):
  print(f"{word1} dan {word2} adalah anagram")
else :
  print(f"{word1} dan {word2} bukan anagram")
```

```python
# PROGRAM TABEL PERKALIAN
angka = int(input("Masukkan angka untuk tabel perkalian: "))

for i in range(1, 11):
  print(f"{angka} x {i} = {angka * i}")
```

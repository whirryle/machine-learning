Read Data Set

```python
import pandas as pd
import matplotlib.pyplot as plt

print(pd.__version__)
```

```python
dskalori = pd.read_excel('kaloriberat.xlsx')
```

```python
dskalori
```

```python
dskalori.info()
```

```python
plt.scatter(dskalori['Kalori'],dskalori['Berat'])
plt.xlabel('Kalori')
plt.ylabel('Berat')
```

INDEPENDENT & DEPENDENT VARIABLE

```python
x = dskalori.iloc[:,[0]].values
y = dskalori.iloc[:,[1]].values
```

```python
x
```

```python
y
```

SPLIT TRAIN TEST

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
regresi = LinearRegression()
```

```python
regresi.fit(x_train,y_train)
```

```python
# SEBARAN DATA TRAIN

plt.scatter(x_train, y_train)

# GARIS REGRESI
plt.plot(x_train,regresi.predict(x_train), c='red')
plt.xlabel('Kalori')
plt.ylabel('Berat')
```

TESTING MODEL

```python
y_prediksi = regresi.predict(x_test)
print(y_prediksi)
```

TEST DATA BARU

```python
new_data = [[410]]
new_prediksi = regresi.predict(new_data)
print(new_prediksi)
```

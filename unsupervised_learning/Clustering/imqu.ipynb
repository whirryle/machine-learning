# IMAGE QUANTIZATION
--> Melakukan kompresi pada gambar dan mengurangi warna menjadi lebih sedikit

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
img = plt.imread("assets/ririn3.jpg")
img.shape
```

```python
plt.imshow(img)
```

#### --- Buat tiap pixel menjadi data ---

```python
img = img.reshape(-1, 3)
img
```

```python
len(img) # total pixel
```

```python
len(np.unique(img, axis=0)) # total unique color
```

#### Cluster 61474 colors into 8 color palette

```python
kmeans = KMeans(n_clusters=8, random_state=42)
labels = kmeans.fit_predict(img)
```

```python
labels
```

```python
palette = kmeans.cluster_centers_.astype(int)
palette
```

```python
quantized_img = palette[labels].reshape(802, 802, 3)
quantized_img
```

```python
plt.imshow(quantized_img)
```

```python
def quantize(img, n_palette):
  n_unique = len(np.unique(img.reshape(-1, 3), axis=0))

  kmeans = KMeans(n_clusters=n_palette, random_state=42)
  labels = kmeans.fit_predict(img.reshape(-1, 3))
  palette = kmeans.cluster_centers_.astype(int)
  quantized_img = palette[labels].reshape(*img.shape)

  plt.figure(figsize=(15, 6))
  plt.subplot(121)
  plt.imshow(img)
  plt.title(f"Original Image [{n_unique} colors]")
  plt.xticks([]); plt.yticks([])
  
  plt.subplot(122)
  plt.imshow(quantized_img)
  plt.title(f"Quantized Image [{n_palette} colors]")
  plt.xticks([]); plt.yticks([])
```

### PERBANDINGAN ANTARA GAMBAR ASLI DAN GAMBAR KUANTISASI


#### Menggunakan 8 color palette

```python
img = plt.imread("assets/ririn3.jpg")
quantize(img, 8)
```

```python
img = plt.imread("assets/cantikkuu.jpeg")
quantize(img, 8)
```
